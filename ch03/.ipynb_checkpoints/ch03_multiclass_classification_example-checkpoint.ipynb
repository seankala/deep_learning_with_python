# 3.5 Classifying Newswires: A Multiclass Classification Example

_In the previous example, we used the IMDb dataset to perform binary classification on movie reviews (i.e. positive review vs. negative review). In this example, we'll use multiclass classification to classify newswires. More specifically, this is an example of single-label multiclass classification._

1. Reuters newswires

2. 46 total classes

## 3.5.1 The Reuters Dataset

```python
from keras.datasets import reuters


(train_data, train_labels), (test_data, test_labels) = reuters.load_data(num_words=10000)

print("Number of training examples: {}".format(len(train_data)))
print("Number of test examples: {}".format(len(test_data)))
```

## 3.5.2 Preparing the Data

```python
import numpy as np


def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.0
        
    return results


x_train, x_test = vectorize_sequences(train_data), vectorize_sequences(test_data)
```

```python
def to_one_hot(labels, dimension=46):
    results = np.zeros((len(labels), dimension))
    
    for i, label in enumerate(labels):
        results[i, label] = 1.0
        
    return results


one_hot_train_labels, one_hot_test_labels = to_one_hot(train_labels), to_one_hot(test_labels)
```

```python
from keras.utils.np_utils import to_categorical


one_hot_train_labels2, one_hot_test_labels2 = to_categorical(train_labels), to_categorical(test_labels)
```

## 3.5.3 Building Your Network

_In the previous example, we used a dense neural network with three layers and the first and second layers having 16 hidden units. This may be okay when there are few labels to classify, but in this case we have 46 and so 16 may be too few hidden units. The units may be prone to dropping or losing information along the process because they act as "information bottlenecks." Therefore, we opt to use 64 hidden units._

```python
from keras import models
from keras import layers


model = models.Sequential()

model.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(46, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
x_val = x_train[:1000]
partial_x_train = x_train[1000:]

y_val = one_hot_train_labels[:1000]
partial_y_train = one_hot_train_labels[1000:]
```

```python
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=20,
                    batch_size=512,
                    validation_data=(x_val, y_val))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plot_graphs(history):
    loss_values = history.history['loss']
    val_loss_values = history.history['val_loss']
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    
    epochs = range(1, len(loss_values) + 1)
    
    fig = plt.figure(figsize=(15, 15))
    ax1 = fig.add_subplot(221)
    plt.grid(True)
    ax2 = fig.add_subplot(222)
    plt.grid(True)
    
    ax1.plot(epochs, loss_values, 'b-', label='Training loss')
    ax1.plot(epochs, val_loss_values, 'r-', label='Validation loss')
    ax1.set_title('Training and Validation Loss')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss')
    handles, labels = ax1.get_legend_handles_labels()
    ax1.legend(handles, labels)
    
    ax2.plot(epochs, acc, 'b-', label='Training acc')
    ax2.plot(epochs, val_acc, 'r-', label='Validation acc')
    ax2.set_title('Training and validation accuracy')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Accuracy')
    handles2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(handles2, labels2)
    
    plt.show()
```

```python
plot_graphs(history)
```

Since we can observe that the model begins to overfit at around 9 epochs, we attempt to build a new model but this time with only 9 epochs.

```python
model_nine = models.Sequential()

model_nine.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model_nine.add(layers.Dense(64, activation='relu'))
model_nine.add(layers.Dense(46, activation='softmax'))

model_nine.compile(optimizer='rmsprop',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

model_nine.fit(partial_x_train,
               partial_y_train,
               epochs=9,
               batch_size=512,
               validation_data=(x_val, y_val))

results = model_nine.evaluate(x_test, one_hot_test_labels)
```

## 3.5.8 Further Experiments

_This time try:_

1. Larger or smaller layers: 32 units, 128 units, etc.

2. Rather than two hidden layers, one and three hidden layers.

```python
# Step 1 : Load data
# ==========================================================================================
from keras.datasets import reuters


(train_data, train_labels), (test_data, test_labels) = reuters.load_data(num_words=10000)


# Step 2 : Prepare data
# ==========================================================================================
import numpy as np
from keras.utils.np_utils import to_categorical


def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.0
        
    return results


x_train, x_test = vectorize_sequences(train_data), vectorize_sequences(test_data)
one_hot_train_labels, one_hot_test_labels = to_categorical(train_labels), to_categorical(test_labels)

x_val = x_train[:1000]
partial_x_train = x_train[1000:]

y_val = one_hot_train_labels[:1000]
pratial_y_train = one_hot_train_labels[1000:]


# Step 3 : Build Network(s)
# ==========================================================================================
from keras import models, layers


# 32 hidden units
model_32 = models.Sequential()
model_32.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_32.add(layers.Dense(32, activation='relu'))
model_32.add(layers.Dense(46, activation='softmax'))

# 128 hidden units
model_128 = models.Sequential()
model_128.add(layers.Dense(128, activation='relu', input_shape=(10000,)))
model_128.add(layers.Dense(128, activation='relu'))
model_128.add(layers.Dense(46, activation='softmax'))

# one hidden layer
model_1l = models.Sequential()
model_1l.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model_1l.add(layers.Dense(46, activation='softmax'))

# three hidden layers
model_3l = models.Sequential()
model_3l.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model_3l.add(layers.Dense(64, activation='relu'))
model_3l.add(layers.Dense(64, activation='relu'))
model_3l.add(layers.Dense(46, activation='softmax'))
```

```python
# Step 4 : Compile and fit models
# ==========================================================================================
model_32.compile(optimizer='rmsprop',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

print("Training model_32...\n")
history_32 = model_32.fit(partial_x_train,
                          partial_y_train,
                          epochs=20,
                          batch_size=512,
                          validation_data=(x_val, y_val))

model_128.compile(optimizer='rmsprop',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

print("Training model_128...\n")
history_128 = model_128.fit(partial_x_train,
                            partial_y_train,
                            epochs=20,
                            batch_size=512,
                            validation_data=(x_val, y_val))

model_1l.compile(optimizer='rmsprop',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])

print("Training model_1l...\n")
history_1l = model_1l.fit(partial_x_train,
                          partial_y_train,
                          epochs=20,
                          batch_size=512,
                          validation_data=(x_val, y_val))

model_3l.compile(optimizer='rmsprop',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])

print("Training model_3l...\n")
history_3l = model_3l.fit(partial_x_train,
                          partial_y_train,
                          epochs=20,
                          batch_size=512,
                          validation_data=(x_val, y_val))
```

## 3.5.9 Wrapping Up


1. The number of units that your last layer should have depends on how many classes you're trying to perform classification on.

2. In a single-label multiclass classification problem, your model should end with a softmax activation function rather than a sigmoid. That way the network will distribute probabilities among all of the relevant classes.

3. Categorical cross-entropy is almost always the loss function that you should use for single-label multiclass classification problems. Essentially, this minimizes the distance between probability distributions output by the network. That way by minimizing this distance we're essentially trying to tune our model to resemble the generalized truth.

4. There are two ways to handle labels in multiclass classification:
  * Encode the labels via categorical encoding (this is provided in `keras.utils.np_utils.to_categorical`) or you could do one-hot encoding yourself.
  * You could encode the labels into integers rather than the one-hot form and then use sparse categorical cross-entropy.
  
5. You should avoid creating networks with hidden units fewer than the number of classes we're trying to classify. This would create information bottlenecks.
